### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

### Read CSV file

```python
df = pd.read_csv('iris_dataset.csv')
```

### List Comprehension for formatting column names from lower and joined with underscore

```python
[" ".join(i) for i in [[i.capitalize() for i in x] for x in [i.split("_") for i in df.columns]]]
```

```python
df.columns = [" ".join(i) for i in [[i.capitalize() for i in x] for x in [i.split("_") for i in df.columns]]]
```

### Scatter Plot

```python
for i in df.columns:
    fig = plt.figure(figsize=(20,15))
    
    for a,y in zip(df.columns,range(1,16)):
        ax = fig.add_subplot(4,4,y)
        if i == a:
            ax.text(0.2, 0.5, i , fontweight="bold", transform=ax.transAxes,fontsize=25)
        else:
            ax.scatter(df[a],df[i])
            ax.set_xlabel(a)
            ax.set_ylabel(i)
    plt.show()
```

### Elbow Curve to get value of k for KMeans

```python
x = df[[i for i in df.columns]]
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),wcss)
```

### K = 2

```python
kmeans = KMeans(2)
kmeans.fit(x)

df['cluster'] = kmeans.fit_predict(x)
df
```

```python
for i in df.columns[:-1]:
    fig = plt.figure(figsize=(20,15))
    
    for a,y in zip(df.columns[:-1],range(1,16)):
        ax = fig.add_subplot(4,4,y)
        if i == a:
            ax.text(0.2, 0.5, i , fontweight="bold", transform=ax.transAxes,fontsize=25)
        else:
            ax.scatter(df[a],df[i],c=df['cluster'],cmap='autumn')
            ax.set_xlabel(a)
            ax.set_ylabel(i)
    plt.show()
```
